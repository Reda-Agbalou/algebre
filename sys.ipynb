### 1  ***Méthode de cholesky***

c'est une méthode directe
>  Il faut s'assurer que la matrice est symétrique définie positive tout d'abord
La première étape consiste à trouver la matrice L de A par cette fonction cholesky() pour en déduire la décomposition de A qui la multiplication de L avec L transposé.

La première étape consiste à trouver la matrice L de A par cette fonction cholesky() pour en déduire la décomposition de A qui la multiplication de L avec L transposé.

```python
import numpy as np
def cholesky(a):
    a=np.array(a,float)
    L=np.zeros_like(a)      #cree matrice de meme ordre que a
    n,_=np.shape(a)     #ordre du matrice carree prenons premier ligne car 2
    for i in range(n):
        for j in range(i,n):
            if j==i:
                L[j,i]=np.sqrt(a[j,i]-np.sum(L[j,:i]**2))    #cours
            else:
                L[j,i]=(a[j,i]-np.sum(L[j,:i]*L[i,:i]))/L[i,i]
    return L
A=[[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]]
b=[1, 0, 1, 0]
L=cholesky(A)
print(L)
```

> La deuxième étape consiste à résoudre le système linéaire Ax=b par la résolution de Ly=b puis L^t x=y

```python
def solveLU(L,U,b):
    L=np.array(L,float)
    U=np.array(U,float)
    b=np.array(b,float)
    n,_=np.shape(L)
    y=np.zeros(n)
    x=np.zeros(n)
    #forward substitution
    for i in range (n):
        sumj=0
        for j in range (i):
            sumj+=L[i,j]*y[j]
        y[i]=(b[i]-sumj)/L[i,i]
    #backward substitution
    for i in range (n-1,-1,-1):
        sumj=0
        for j in range (i+1,n):
            sumj+=U[i,j]*x[j]
        x[i]=(y[i]-sumj)/U[i,i]
    return x
X=solveLU(L,np.transpose(L),b)
print(f"la solution de ce système par la méthode de cholesky est : {X}")
```

</p>

### 3 ***Méthode de Jacobi :***

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
La méthode de Jacobi est une méthode itérative pour résoudre un système linéaire Ax = b. Elle consiste à itérer sur chaque équation du système et à mettre à jour la valeur de chaque inconnue en fonction des autres inconnues déjà connues.
<p>

```python
import numpy as np

def Jacobi(A, b, x0, tol, maxiter):
    """
    Solve a system of linear equations Ax = b using the Jacobi method.

    Parameters:
    - A: coefficient matrix (n x n)
    - b: right-hand side vector (n x 1)
    - x0: initial guess for the solution (n x 1)
    - tol: tolerance for the residual (scalar)
    - maxiter: maximum number of iterations (scalar)

    Returns:
    - x: the solution vector (n x 1)
    - iterations: the number of iterations performed (scalar)
    - res: the final residual (scalar)
    """

    # Get the dimensions of the matrix
    n = A.shape[0]

    # Initialize the iteration counter
    iterations = 0

    # Compute the initial residual
    res = np.linalg.norm(b - np.matmul(A, x0))

    # Initialize the solution vector
    x = x0

    # Start the iteration loop
    while iterations < maxiter and res > tol:
        # Increment the iteration counter
        iterations += 1

        # Loop over the rows of the matrix
        for i in range(n):
            # Compute the new value for the i-th component of the solution vector
            x[i] = (b[i] - np.dot(A[i, :i], x[:i]) - np.dot(A[i, i + 1:], x[i + 1:])) / A[i, i]

        # Compute the new residual
        res = np.linalg.norm(b - np.matmul(A, x))

    # Return the solution vector, the number of iterations performed, and the final residual
    return x, iterations, res

A = np.array([[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]
])
b = np.array([1, 0, 1, 0])
x0 = np.zeros(4)  # initial guess for the solution
tol = 1e-6  # tolerance for the residual
maxiter = 100  # maximum number of iterations

x, iterations, res = Jacobi(A, b, x0, tol, maxiter)

print(x)
print(iterations)
print(res)
```

### 4 ***Méthode de gradient à pas optimal   :***

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
    Resoudre AX=b par la methode du gradient à pas optimal 
    <br>
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    Principle : <br>
        We consider the following problem: minimize on $R^{n}$ the function $\phi$ defined by
    $$
    \phi(x)=\frac{1}{2} x^{T} A x-b^{T} x
    $$
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">  On a A est une matrice symetrique definie positive.
    The interest of the question comes from the fact that the minimum of $\phi$ is reached for $x^{*}=A^{-1} b$, and this solution is unique. Indeed, the gradient of $\phi$ in $x=\left(x_{1} \ldots x_{n}\right)^{T} \in R^{n}$ is the vector:
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    $$
    \nabla \phi(x)=\left(\frac{\partial \phi}{\partial x_{1}} \frac{\partial \phi}{\partial x_{2}} \cdots \frac{\partial \phi}{\partial x_{n}}\right)^{T}=\frac{1}{2} A x+\frac{1}{2} A^{T} x-b=A x-b
    $$
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    which vanishes only for $x^{*}=A^{-1} b$. Note that in this calculation we have used the fact that $A$ is a symmetric matrix. The vector $x^{*}=A^{-1} b$ is therefore the unique critical point of $\phi$ and, since $A$ is positive definite, it is a global minimum.
    <br>
</p>

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
    To calculate $x^{(k+1)}$ from $x^{(k)}$, we then minimize the function $\phi$ on the direction vector line $p^{(k)}$ and going through $x^{(k)}$: we therefore choose
    $$
    x^{(k+1)}=x^{(k)}+\alpha_{k} p^{(k)}
    $$
    where $\alpha_{k} \in R$
    <br>
</p>

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
    To calculate $x^{(k+1)}$ from $x^{(k)}$, we then minimize the function $\phi$ on the direction vector line $p^{(k)}$ and going through $x^{(k)}$: we therefore choose
    $$
    x^{(k+1)}=x^{(k)}+\alpha_{k} p^{(k)}
    $$
    where $\alpha_{k} \in R$
    <br>
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    definition of the function $\phi(x)$ and $\nabla \phi(x)$
    <br>
</p>

```python
def phi(A, b, x):
    return 0.5 * np.dot(np.dot(np.transpose(x), A), x) - np.dot(np.transpose(b), x)
```

```python
def Dphi(A, b, x):
    return np.dot(A, x) - b
```

```python
A = np.array([[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]])
B = np.array([1, -1, 0, 1])
n = len(A)
Alpha = 0.01  # step
epsilon = 1e-9  # convergence tolerance
max_iters = 1e6  # maximum iterations

x0 = np.array([1, 0, 0, 1])
```

```python
# Gradient method function
def gradient_methode(A, b, x0):
    # initialization of the starting vector
    next_x = x0
    # Display initial vector
    print("i = 0 ; phi(x)= " + str(phi(A, b, next_x)))
    i = 1
    # convergence flag
    cvg = False
    print("Search start")
    while i <= max_iters:
        curr_x = next_x
        next_x = curr_x - Alpha * Dphi(
            A, b, curr_x
        )  # formula of X(k+1) under the gradient method

        diff = next_x - curr_x
        # convergence test
        if (
            np.linalg.norm(diff, 2) / (np.linalg.norm(next_x, 2) + np.finfo(float).eps)
            <= epsilon
        ):
            cvg = True
            break
        # display the Approximate value of phi(x) during the search for the minimum
        print("i = " + str(i) + " ; phi(x)= " + str(phi(A, b, next_x)))
        i += 1

    # if we have convergence we display the solution as result if not we display a message
    # that the sequence is divergent
    if cvg:
        print("The Min obtained in " + str(i) + " iterations.")
        print("x_sol =", next_x)
    else:
        print("the sequence isn't converge.")

    return next_x
```

```python
x_sol = gradient_methode(A, B, x0)
```

```python
print("The solution by the gradient method:")
print("X =  ",x_sol)
```

### 5 ***Calcule des valeurs propre par la methode de Givens   :***

La méthode de Givens est une techniques utilisée pour factoriser une matrice en produit de matrices triangulaires et orthogonales. 

```python
import numpy as np

# Matrice A
A = np.array([[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]])

# Fonction de rotation de Givens
def givens_rotation(a, b):
  if b == 0:
    c = 1
    s = 0
  else:
    if abs(b) > abs(a):
      tau = -a / b
      s = 1 / np.sqrt(1 + tau**2)
      c = s * tau
    else:
      tau = -b / a
      c = 1 / np.sqrt(1 + tau**2)
      s = c * tau
  return c, s

# Méthode de Givens-Householder
def givens_householder(A):
  m, n = A.shape
  Q = np.eye(m)
  R = A.copy()
  for j in range(n):
    for i in range(m-1, j, -1):
      c, s = givens_rotation(R[i-1, j], R[i, j])
      G = np.eye(m)
      G[i-1, i-1] = c
      G[i, i] = c
      G[i-1, i] = s
      G[i, i-1] = -s
      R = G.dot(R)
      Q = Q.dot(G.T)
  return Q, R

# Calcul des valeurs propres
Q, R = givens_householder(A)
vals, vecs = np.linalg.eig(R)

# Valeurs propres triées
vals = np.sort(vals)

print(vals)
```

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
    La méthode de Newton est une méthode itérative pour résoudre des équations non linéaires.La méthode de Newton est souvent utilisée pour trouver les racines d'une fonction, c'est-à-dire les points où la fonction s'annule.
    Dans l'exercice suivant on va resoudre le Systeme suivant avec la methode de Newton <br>
    the system concerned is as follows:<br>
$$
\left\{
    \begin{array}\\
        x^{2}-y = 0 \\
        \               &avec (x_0,y_0)=(1,1) \\
        y^{2}=0 \\
    \end{array}
\right.
$$
</p>


définition des fonctions du système


```python
def f1(x, y):
    return x ** 2 - y


def f2(x, y):
    return y ** 2
```

définition de la jacobienne


```python
def jacobian(x, y):
    return np.array([[2 * x, -1], [0, 2 * y]])
```

```python
# définition de la méthode de Newton
def newton_method(x0, y0, epsilon=1e-7, max_iter=100):
    # initialisation des variables
    x = x0
    y = y0
    converged = False
    iter = 0

    # boucle jusqu'à convergence ou nombre maximal d'itérations atteint
    while not converged and iter < max_iter:
        # calcul de la jacobienne en (x, y)
        J = jacobian(x, y)
        # calcul de l'inverse de la jacobienne
        Jinv = np.linalg.inv(J)
        # calcul de F(x, y) et F(x0, y0)
        F = np.array([f1(x, y), f2(x, y)])
        F0 = np.array([f1(x0, y0), f2(x0, y0)])
        # calcul de la différence entre F(x, y) et F(x0, y0)
        deltaF = F - F0
        # calcul de la différence entre (x, y) et (x0, y0)
        delta = Jinv.dot(deltaF)
        # mise à jour des estimations de x et y
        x = x0 + delta[0]
        y = y0 + delta[1]
        # vérification de la convergence
        if np.abs(delta).max() < epsilon:
            converged = True
        # incrémentation du compteur d'itérations
        iter += 1

    # retourne la solution trouvée et le nombre d'itérations
    return (x, y), iter


# test de la méthode en utilisant (x0, y0) = (1, 1) comme point de départ
solution, nb_iter = newton_method(1, 1)
print("Solution trouvée:", solution)
print("Nombre d'itérations:", nb_iter)
```
